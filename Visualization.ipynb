```python
import pandas as pd
import numpy as np

```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
import seaborn as sns
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)

```

```python
num = np.random.normal(1,10,1000)
sns.displot(num, kde=True) # Histogram with a kernel density estimate (Density Curve)
plt.show()
```

```python
sns.displot(num, kde=False, rug= True)
```

```python
housing = pd.read_excel('https://github.com/dayanandv/Data-Science/raw/main/dataset/Housing_data.xlsx')
housing.hist(bins=40, figsize=(20,20))
plt.show()
```

```python
exercise =pd.read_csv('https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/exercise.csv')
sns.swarmplot(x='kind',y='pulse',hue='diet', size = 10, data =exercise)
plt.show
```

```python
plt.figure(figsize=(7,7))
sns.swarmplot(x= "kind", y = "pulse", hue="diet", size=9, data = exercise)
plt.show()


```

# Box Plot


```python
insurance = pd.read_csv('https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/insurance.csv')
insurance.head()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(x=insurance.smoker, y=insurance.charges)
plt.show()
```
